```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Given data
city_levels = ['First-tier', 'New first-tier', 'Second-tier', 'Third-tier', 'Fourth-tier', 'Fifth-tier']
percentage_distribution = [27.87, 25.87, 19.73, 16.80, 8.27, 7.47]

# Create a bar chart
plt.bar(city_levels, percentage_distribution, color='skyblue')
plt.title('City Level vs. Psychological Institution Distribution Percentage')
plt.xlabel('City Level')
plt.ylabel('Percentage')
plt.ylim(0, 30)  # Set the y-axis limit for better visualization

# Add data labels to the bars
for i, percentage in enumerate(percentage_distribution):
    plt.text(i, percentage + 0.2, f'{percentage:.2f}%', ha='center')

# Show or save the chart
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\city_level_distribution.png', dpi=1200)
plt.show()
```
