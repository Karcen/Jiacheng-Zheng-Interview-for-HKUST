```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Data
labels = ['Excellent Mental Health', 'Moderate Mental Health', 'Poor Mental Health', 'Uncertain Mental Health']
sizes = [35.90, 48.10, 13.90, 2.10]
colors = ['lightcoral', 'lightblue', 'lightgreen', 'lightskyblue']

# Plotting the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Set chart title
plt.title('Distribution of Mental Health Conditions')

# Display the pie chart
plt.axis('equal')  # Maintain equal aspect ratio for the pie chart
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Distribution of Mental Health Conditions.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
