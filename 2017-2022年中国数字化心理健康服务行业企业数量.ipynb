```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Garamond'
# 年份数据
years = [2017, 2018, 2019, 2020, 2021, 2022]

# 企业数量数据
enterprise_count = [220, 290, 375, 402, 420, 437]

# 新增数量数据
new_count = [80, 80, 70, 30, 6, 18]

# 将中文转换为英文
years_label = ['2017', '2018', '2019', '2020', '2021', '2022']
enterprise_label = 'Enterprise Count'
new_label = 'New Count'

# 设置每个年份的索引位置
ind = np.arange(len(years))

# 设置图形大小
plt.figure(figsize=(10, 6))

# 绘制企业数量柱状图
plt.bar(ind - 0.2, enterprise_count, 0.4, color='#E6CEAC', label=enterprise_label)

# 绘制新增数量柱状图
plt.bar(ind + 0.2, new_count, 0.4, color='#ECAD9E', label=new_label)

# 设置横轴标签和刻度
plt.xticks(ind, years_label)

# 添加图例
plt.legend()

# 设置纵轴标签
plt.ylabel('Count')

# 标题
plt.title('Number of Enterprises in China\'s Digital Mental Health Service Industry From 2017 To 2022')

plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Number of Enterprises in Chinas Digital Mental Health Service Industry From 2017 To 2022_en.png', dpi=1200, bbox_inches = 'tight')
# 显示柱状图
plt.show()
```
