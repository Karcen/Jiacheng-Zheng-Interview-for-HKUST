```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties


font = FontProperties(fname=r"C:\Windows\Fonts\simhei.ttf", size=12)


categories = ['总覆盖排放', '一带一路覆盖排放', '非一带一路覆盖排放']
imports_excluded_5 = [2.67, 2.25, 0.42]
imports_excluded_10 = [1.54, 1.35, 0.19]
imports_not_excluded = [1.2, 1.16, 0.05]


bar_width = 0.15
index = range(len(categories))

fig, ax = plt.subplots()

bar1 = ax.bar(index, imports_excluded_5, bar_width, label='进口排除<5%')
bar2 = ax.bar([i + bar_width for i in index], imports_excluded_10, bar_width, label='进口排除<10%')
bar3 = ax.bar([i + 2 * bar_width for i in index], imports_not_excluded, bar_width, label='无进口排除')


# ax.set_xlabel('排放类型', fontproperties=font)
ax.set_ylabel('排放量($CO_2\quad Gt$)', fontproperties=font, fontsize=12)
# ax.set_title('不同排放类型的排放量比较', fontproperties=font)
ax.set_xticks([i + bar_width for i in index])
ax.set_xticklabels(categories, fontproperties=font, fontsize=12)
ax.legend(prop=font)
plt.tight_layout()
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\figure2.png', dpi=1200)


plt.show()
```
