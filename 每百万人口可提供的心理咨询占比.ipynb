```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
countries = ['China', 'United States']
proportions = [20, 1000]

plt.bar(countries, proportions, color='orange')

# Add labels to the bars
for i in range(len(countries)):
    plt.text(countries[i], proportions[i], str(proportions[i]), ha='center', va='bottom')

plt.xlabel('Countries')
plt.ylabel('Proportion per Million')
plt.title('Proportion of Mental Counseling per Million Population')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Proportion of Mental Counseling per Million Population_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
