```python
import matplotlib.pyplot as plt

# Data
channels = ['Others', 'Baidu Search', 'Friends Referral', 'Bilibili/Weibo/Xiaohongshu/Zhihu', 'Psychology Books, Courses & Lectures', 'Personal Short Videos', 'Psychological Institution Short Videos', 'Psychology Official Accounts']
proportions = [9.30, 33.54, 17.07, 44.36, 38.57, 34.30, 26.37, 70.58]

# Create horizontal bar chart
plt.barh(channels, proportions)  # Use plt.barh() for horizontal bar chart

# Add title and labels
plt.title('Channels for Patients to Understand the Knowledge of Depression')
plt.xlabel('Proportions')
plt.ylabel('Channels')

# Add labels
for i in range(len(channels)):
    plt.text(proportions[i], channels[i], f"{proportions[i]}%", ha='left', va='center')

# Show the chart
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Channels_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
