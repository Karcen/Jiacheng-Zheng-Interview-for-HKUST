```python
import matplotlib.pyplot as plt
import random

# 随机生成10个在[20, 100]之间的整数
radii = [random.randint(20, 100) for _ in range(10)]

# 颜色列表
colors = ['red', 'green', 'blue', 'purple']

# 创建绘图
fig, ax = plt.subplots()
ax.set_aspect('equal', 'box')  # 确保圆形比例

# 绘制同心圆
for radius in radii:
    color = random.choice(colors)  # 随机选择颜色
    circle = plt.Circle((0, 0), radius, color=color, fill=False, linewidth=2)
    ax.add_patch(circle)

# 设置坐标轴范围
ax.set_xlim(-110, 110)
ax.set_ylim(-110, 110)

# 隐藏坐标轴
ax.axis('off')

# 显示绘图
plt.show()
```

```python
import random

# 生成0-100的随机整数
target = random.randint(0, 100)
attempts = 0  # 记录用户猜测次数

print("猜数字游戏：请输入0到100之间的整数！")

# 循环直到用户猜中
while True:
    guess = int(input("请输入你猜的数字："))
    attempts += 1  # 每次输入后增加一次尝试计数
    
    if guess > target:
        print("太大了！")
    elif guess < target:
        print("太小了！")
    else:
        print(f"恭喜你，猜中了！这个数字是 {target}。")
        print(f"你一共猜了 {attempts} 次。")
        break
```
