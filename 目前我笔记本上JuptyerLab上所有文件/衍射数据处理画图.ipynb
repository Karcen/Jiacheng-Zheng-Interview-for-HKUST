```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Set the font properties
plt.rcParams['font.family'] = 'Garamond'
plt.rcParams['font.style'] = 'normal'

# Read the Excel file (replace 'your_file.xlsx' with your file name)
file_path = r'C:\Users\Jiacheng Zheng\Downloads\è¡å°.xlsx'  # Replace with your file path
data = pd.read_excel(file_path, sheet_name='Sheet2')

# Check if the headers are correct
print(data.head())

# Extract position and power meter readings
positions = data['Position/mm']
intensity = data['Power Meter Reading/Î¼W']

# Calculate relative intensity
relative_intensity = intensity / intensity.max()

# Determine the central bright fringe position
central_position = positions[relative_intensity.idxmax()]

# Format x-axis labels to show 3 decimal places
def format_func(value, tick_number):
    return f'{value:.3f}'

# Plot the graph with the central position marked
plt.figure(figsize=(8, 6))
plt.plot(positions, relative_intensity, label="Relative Intensity")
plt.axvline(x=central_position, color='red', linestyle='--', label='Central Bright Fringe')

# Set labels and title
plt.xlabel('x/(mm)', size=15)
plt.ylabel('I/$I_0$', size=15)

# Format the x-axis to show 3 decimal places
plt.gca().xaxis.set_major_formatter(FuncFormatter(format_func))

# Show legend and grid
plt.legend()
plt.grid(True)

# Adjust layout
plt.tight_layout()

# Save the graph or display it
plt.savefig(r'C:\Users\Jiacheng Zheng\Downloads\diffraction_plot.svg')  # Save as an image file
plt.show()
```
