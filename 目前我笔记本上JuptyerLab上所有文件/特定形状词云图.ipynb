```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image

# 读取 Excel 文件
df = pd.read_excel(r'C:\Users\Jiacheng Zheng\Desktop\data.xlsx')

# 获取第 E 列数据
column_E = df['Authors']

# 将所有文本连接成一个字符串
text = ' '.join(column_E.dropna().astype(str))

# 分割文本成单词并排除单个字母和少于三个字母的单词
words = text.split()
filtered_words = [word for word in words if len(word) > 3]

# 计算单词频率
word_counts = Counter(filtered_words)

# 输出具体数据
print("单词频率：")
for word, count in word_counts.most_common():
    print(f"{word}: {count}")

# 生成词云图
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_counts)

# 绘制词云图
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# 保存词云图为png
wordcloud.to_file(r"C:\Users\Jiacheng Zheng\Desktop\wordcloud.png")


```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# 读取 Excel 文件
df = pd.read_excel(r'C:\Users\Jiacheng Zheng\Desktop\data.xlsx')

# 获取第 E 列数据
column_E = df['Authors']

# 将所有文本连接成一个字符串
text = ' '.join(column_E.dropna().astype(str))

# 分割文本成单词并排除单个字母和少于三个字母的单词
words = text.split()
filtered_words = [word for word in words if len(word) > 3]

# 计算单词频率
word_counts = Counter(filtered_words)

# 输出具体数据
print("单词频率：")
for word, count in word_counts.most_common():
    print(f"{word}: {count}")

# 创建 "MLN" 形状数组
mask = np.array(Image.open(r'C:\Users\Jiacheng Zheng\Desktop\MYN.jpg'))

# 生成词云图，指定形状
wordcloud = WordCloud(width=800, height=400, background_color='white', mask=mask).generate_from_frequencies(word_counts)

# 绘制词云图
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# 保存词云图为png
wordcloud.to_file(r"C:\Users\Jiacheng Zheng\Desktop\wordcloud.png")
```
