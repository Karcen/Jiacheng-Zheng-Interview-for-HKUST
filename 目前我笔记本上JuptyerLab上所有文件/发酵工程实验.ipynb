```python
import pandas as pd
from scipy.stats import kruskal

# Load your data into pandas DataFrame
data = {
    'KCl_concentration': [0, 0.05, 0.10, 0.15, 0.20],
    'KCl_avg_dry_weight': [0.06, 0.05, 0.08, 0.04, 0.07],
    'K2HPO4_concentration': [0, 0.10, 0.20, 0.30, 0.40],
    'K2HPO4_avg_dry_weight': [0.05, 0.05, 0.11, 0.05, 0.14],
    'original_medium_avg_dry_weight': [0.07, 0.05, 0.11, pd.NA, 0.08]
}

df = pd.DataFrame(data)

# Perform Kruskal-Wallis test for KCl concentration
_, p_value_kcl = kruskal(df['KCl_avg_dry_weight'], df['K2HPO4_avg_dry_weight'], df['original_medium_avg_dry_weight'].dropna())

# Print result
print("KCl Kruskal-Wallis p-value:", p_value_kcl)

# Perform Kruskal-Wallis test for K2HPO4 concentration
_, p_value_k2hpo4 = kruskal(df['K2HPO4_avg_dry_weight'], df['original_medium_avg_dry_weight'].dropna())

# Print result
print("K2HPO4 Kruskal-Wallis p-value:", p_value_k2hpo4)
```

```python
import matplotlib.pyplot as plt

# KCl data
KCl_concentration = [0, 0.05, 0.10, 0.15, 0.20]
KCl_avg_dry_weight = [0.06, 0.05, 0.08, 0.04, 0.07]

# K2HPO4 data
K2HPO4_concentration = [0, 0.10, 0.20, 0.30, 0.40]
K2HPO4_avg_dry_weight = [0.05, 0.05, 0.11, 0.05, 0.14]

# Plotting the curve
plt.figure(figsize=(10, 6))

# Plotting KCl curve
plt.plot(KCl_concentration, KCl_avg_dry_weight, marker='o', linestyle='-', color='b', label='KCl')

# Plotting K2HPO4 curve
plt.plot(K2HPO4_concentration, K2HPO4_avg_dry_weight, marker='o', linestyle='-', color='r', label='K2HPO4')

# Setting up the plot properties
plt.title('Curve of KCl and K2HPO4 Concentration vs. Average Dry Weight')
plt.xlabel('Concentration (g/L)')
plt.ylabel('Average Dry Weight (g/L)')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline

# KCl data
KCl_concentration = [0, 0.05, 0.10, 0.15, 0.20]
KCl_avg_dry_weight = [0.06, 0.05, 0.08, 0.04, 0.07]

# K2HPO4 data
K2HPO4_concentration = [0, 0.10, 0.20, 0.30, 0.40]
K2HPO4_avg_dry_weight = [0.05, 0.05, 0.11, 0.05, 0.14]

# Creating new x-values for smoother curves
x_smooth = np.linspace(min(KCl_concentration + K2HPO4_concentration), max(KCl_concentration + K2HPO4_concentration), 300)

# Smoothing KCl curve
spl_KCl = make_interp_spline(KCl_concentration, KCl_avg_dry_weight)
KCl_smooth = spl_KCl(x_smooth)

# Smoothing K2HPO4 curve
spl_K2HPO4 = make_interp_spline(K2HPO4_concentration, K2HPO4_avg_dry_weight)
K2HPO4_smooth = spl_K2HPO4(x_smooth)

# Plotting the smoothed curves
plt.figure(figsize=(10, 6))

# Plotting smoothed KCl curve
plt.plot(x_smooth, KCl_smooth, color='b', label='KCl')

# Plotting smoothed K2HPO4 curve
plt.plot(x_smooth, K2HPO4_smooth, color='r', label='K2HPO4')

# Setting up the plot properties with Garamond font
plt.title('Smoothed Curve of KCl and K$_2$HPO$_4$ Concentration vs. Average Dry Weight', fontname='Garamond', fontsize=16)
plt.xlabel('Concentration (g/L)', fontname='Garamond', fontsize=14)
plt.ylabel('Average Dry Weight (g/L)', fontname='Garamond', fontsize=14)
plt.grid(True)
plt.legend(prop={'family': 'Garamond', 'size': 12})

# Set Garamond font for tick labels
plt.xticks(fontname='Garamond', fontsize=12)
plt.yticks(fontname='Garamond', fontsize=12)

# Save the plot as a PNG image
plt.savefig('KCl_K$_2$HPO$_4$_concentration_vs_avg_dry_weight.png', dpi=1200, bbox_inches='tight')

# Display the plot
plt.show()
```
