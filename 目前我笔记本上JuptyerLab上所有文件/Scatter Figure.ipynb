```python
import pandas as pd
import matplotlib.pyplot as plt

# 设置字体为 Garamond
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = 'Garamond'

# 读取第二个Excel文件的第一个工作表，范围A2:C46
df3 = pd.read_excel(r'E:\China_vs_India\China\National\CAUSITIVE.xlsx', sheet_name='1996-2006', usecols='A:C', skiprows=1, nrows=45)
# print(df.head())
# 提取第三个需要绘制的数据列
y3 = df3['2001Diag']
x3 = df3['2001ODE']
sectors3 = df3['Sector']

# 创建第三个散点图
plt.figure(figsize=(8, 6))  # 设置图像大小为 8x6
plt.subplot(2, 2, 3)
plt.scatter(x3, y3)
plt.xlabel('2001ODE')
plt.ylabel('2001Diag')
plt.title('Input-Output Structural Change: 2009-2019')
# 添加y=1的标注线
plt.axhline(y=1, color='r', linestyle='--', linewidth=1)

# 添加x=0的标注线
plt.axvline(x=0, color='b', linestyle='--', linewidth=1)

# 显示每个数据点的标记
for i, txt in enumerate(sectors3):
    plt.annotate(txt, (x3[i], y3[i]))

# 读取第二个Excel文件的第一个工作表，范围F2:H46
df4 = pd.read_excel(r'E:\China_vs_India\China\National\CAUSITIVE.xlsx', sheet_name='1996-2006', usecols='F:H', skiprows=1, nrows=45)
# print(df.head())
# 提取第四个需要绘制的数据列
y4 = df4['2006Diag']
x4 = df4['2006ODE']
sectors4 = df4['Sector.1']
# 创建第四个散点图
plt.subplot(2, 2, 4)
plt.scatter(x4, y4)
plt.xlabel('2006ODE')
plt.ylabel('2006Diag')
plt.title('Input-Output Structural Change: 2009-2019')

# 添加y=1的标注线
plt.axhline(y=1, color='r', linestyle='--', linewidth=1)

# 添加x=0的标注线
plt.axvline(x=0, color='b', linestyle='--', linewidth=1)

# 显示每个数据点的标记
for i, txt in enumerate(sectors4):
    plt.annotate(txt, (x4[i], y4[i]), fontsize=8)


plt.tight_layout()
# plt.savefig('scatter_plot.svg', format='svg')
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# 读取第一个Excel文件的第一个工作表，范围A2:C46
df1 = pd.read_excel(r'E:\China_vs_India\China\National\CAUSITIVE.xlsx', sheet_name='1996-2006', usecols='A:C', skiprows=1, nrows=45)
# print(df1.head())
# 绘制第一个散点图
fig1 = px.scatter(df1, x='2001ODE', y='2001Diag', title='Input-Output Structural Change: 1996-2006')

# 在y=1处添加标记线
fig1.add_shape(type="line", x0=fig1.data[0].x.min(), y0=1, x1=fig1.data[0].x.max(), y1=1, line=dict(color="red", width=1, dash="dash"))

# 在x=0处添加标记线
fig1.add_shape(type="line", x0=0, y0=fig1.data[0].y.min(), x1=0, y1=fig1.data[0].y.max(), line=dict(color="blue", width=1, dash="dash"))

# 显示每个数据点的标签
fig1.update_traces(text=df1['Sector'])

# 保存为HTML文件
# fig1.write_html("scatter_plot_1996-2006_1.html")


# 读取第二个Excel文件的第一个工作表，范围F2:H46
df2 = pd.read_excel(r'E:\China_vs_India\China\National\CAUSITIVE.xlsx', sheet_name='1996-2006', usecols='F:H', skiprows=1, nrows=45)
# print(df2.head())
# 绘制第二个散点图
fig2 = px.scatter(df2, x='2006ODE', y='2006Diag', title='Input-Output Structural Change: 1996-2006')

# 在y=1处添加标记线
fig2.add_shape(type="line", x0=fig2.data[0].x.min(), y0=1, x1=fig2.data[0].x.max(), y1=1, line=dict(color="red", width=1, dash="dash"))

# 在x=0处添加标记线
fig2.add_shape(type="line", x0=0, y0=fig2.data[0].y.min(), x1=0, y1=fig2.data[0].y.max(), line=dict(color="blue", width=1, dash="dash"))

# 显示每个数据点的标签
fig2.update_traces(text=df2['Sector.1'])

# 保存为HTML文件
# fig2.write_html("scatter_plot_1996-2006_2.html")

# 显示两个图
fig1.show()
fig2.show()
```
