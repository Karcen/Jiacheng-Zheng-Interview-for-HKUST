```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Data
marital_status = ['Unmarried, No Partner', 'Unmarried, Has Partner', 'Married', 'Divorced']
risk_probability = [23.60, 13.80, 5.10, 11.70]

# Plotting
plt.barh(marital_status, risk_probability)
plt.ylabel('Marital Status')
plt.xlabel('Risk Probability (%)')
plt.title('Relationship between Marital Status and Risk of Depression')
plt.legend(['Risk Probability'])

# Add labels on right side of each bar
for i, v in enumerate(risk_probability):
    plt.text(v, i, str(v)+'%', ha='left', va='center')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Relationship between Marital Status and Risk of Depression_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
