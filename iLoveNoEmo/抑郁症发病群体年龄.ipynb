```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = {"18岁以下": 30.28, "18~24岁": 35.32, "25~30岁": 16.82, "31~40岁": 13.00, "41~50岁": 2.75, "51~60岁": 1.83, "61岁及以上": 0.50}
# 构造数据
data = pd.DataFrame([data])
# 绘制圆环图，并返回饼块对象
wedges, texts = plt.pie(data.iloc[0], wedgeprops={"width": 0.5})
# 构造annotate函数的**kw参数，设置引导线线型
kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")
# 遍历饼块绘制注释标签和引导线
for i, p in enumerate(wedges):
    # 根据matplotlib.patches.Wedge对象的theta1和theta2参数计算饼块均分点的角度
    ang = (p.theta2 - p.theta1) / 2.0 + p.theta1
    # 根据角度的弧度计算 饼块均分点的坐标（引导线的起点）
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    # 根据x的值即角度所在象限确定引导线的对齐方式
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    # 设置引导线的连接方式
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    # 绘制注释标签和引导线
    plt.annotate(
        f"{data.columns[i]}: {data.iloc[0, i]}%",  # 添加数据
        xy=(x, y),
        xytext=(1.35 * np.sign(x), 1.4 * y),
        horizontalalignment=horizontalalignment,
        **kw
    )

plt.title("抑郁症发病群体年龄")
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\age_Prop.png', dpi=1200)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Garamond'

data = {"Under 18 years old": 30.28, "18-24 years old": 35.32, "25-30 years old": 16.82, "31-40 years old": 13.00, "41-50 years old": 2.75, "51-60 years old": 1.83, "Over 61 years old": 0.50}
# Create data frame
data = pd.DataFrame([data])
# Draw a pie chart and return wedges and texts objects
wedges, texts = plt.pie(data.iloc[0], wedgeprops={"width": 0.5})
# Set up **kw parameters of annotate function to set the style of guide lines
kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")
# Iterate through wedges to draw labels and guide lines
for i, p in enumerate(wedges):
    # Calculate the angle of the pie slice at the midpoint
    ang = (p.theta2 - p.theta1) / 2.0 + p.theta1
    # Calculate the coordinates of the midpoint (starting point of the guide line) based on the angle in radians
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    # Determine the alignment of the guide line based on the value of x (quadrant of the angle)
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    # Set up the connection style of the guide line
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    # Draw the label and guide line
    plt.annotate(
        f"{data.columns[i]}: {data.iloc[0, i]}%",  # Add data
        xy=(x, y),
        xytext=(1.35 * np.sign(x), 1.4 * y),
        horizontalalignment=horizontalalignment,
        **kw
    )

plt.title("Age Distribution of Depression Patients")
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\age_Prop_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
