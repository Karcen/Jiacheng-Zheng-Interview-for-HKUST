```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Garamond'

# Data
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
num_of_counseling_agencies = [1240, 1405, 2128, 3759, 5305, 7744, 10519, 13984, 19678, 15955, 18844]

# Plotting the bar chart
bars = plt.bar(years, num_of_counseling_agencies, color='skyblue')

# Adding data labels on top of each bar
for bar, label in zip(bars, num_of_counseling_agencies):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{int(label)}', 
             ha='center', va='bottom', fontsize=8, color='black')

# Set chart title and labels
plt.title('Number of Counseling Agencies Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Counseling Agencies')

# Save the chart as an image
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Number_of_Counseling_Agencies_Over_the_Years_en.png', dpi=1200, bbox_inches='tight')

# Display the bar chart
plt.show()
```
