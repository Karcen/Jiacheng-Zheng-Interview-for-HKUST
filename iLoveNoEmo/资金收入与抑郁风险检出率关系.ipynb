```python
import matplotlib.pyplot as plt

# Data
plt.rcParams['font.family'] = 'Garamond'
income_range = ['Above 10,000 RMB', '6,000 - 10,000 RMB', '4,000 - 6,000 RMB', '2,000 - 4,000 RMB', 'Below 2,000 RMB']
high_risk_rate = [91.30, 93.10, 90.20, 86.90, 79.00]
low_risk_rate = [8.70, 6.90, 9.80, 13.10, 21.00]

# Calculate cumulative percentages
high_risk_cumulative = [sum(high_risk_rate[:i+1]) for i in range(len(income_range))]
low_risk_cumulative = [sum(low_risk_rate[:i+1]) for i in range(len(income_range))]

# Plotting
fig, ax = plt.subplots()
bar_width = 0.35
index = range(len(income_range))

high_risk_bar = ax.barh(index, high_risk_cumulative, bar_width, label='High Risk Rate')
low_risk_bar = ax.barh(index, low_risk_cumulative, bar_width, label='Low Risk Rate')

# Add labels in the middle of each bar
for i, v in enumerate(high_risk_cumulative):
    ax.text(v/2, i, f'{high_risk_rate[i]}%', va='center')

for i, v in enumerate(low_risk_cumulative):
    ax.text(v/2, i, f'{low_risk_rate[i]}%', va='center')

# Set legend and title
plt.xlabel('Cumulative Detection Rate of Depression Risk')
plt.ylabel('Income Range')
plt.title('Relationship between Income Range and Detection Rate of Depression Risk')
plt.legend()
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Relationship between Income Range and Detection Rate of Depression Risk_en.png', dpi=1200, bbox_inches = 'tight')
# Display the plot
plt.show()
```
