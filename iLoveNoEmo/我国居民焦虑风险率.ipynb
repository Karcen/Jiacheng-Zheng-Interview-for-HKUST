```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Data
age_groups = ['55 and above', '45-54', '35-44', '25-34', '18-24']
low_risk = [93.70, 94.30, 95.30, 87.70, 75.90]
high_risk = [6.30, 5.70, 4.70, 12.30, 24.10]

# Calculate cumulative percentages
low_risk_cumulative = [sum(low_risk[:i+1]) for i in range(len(age_groups))]
high_risk_cumulative = [sum(high_risk[:i+1]) for i in range(len(age_groups))]

# Plotting
plt.bar(age_groups, low_risk_cumulative, label='Low risk rate')
plt.bar(age_groups, high_risk_cumulative, bottom=low_risk_cumulative, label='High risk rate')

# Add labels
for i in range(len(age_groups)):
    plt.text(i, low_risk_cumulative[i]/2, f'{low_risk[i]}%', ha='center', color='black')
    plt.text(i, low_risk_cumulative[i]+high_risk_cumulative[i]/2, f'{high_risk[i]}%', ha='center', color='black')

# Set legend and title
plt.legend()
plt.xlabel('Age')
plt.ylabel('Cumulative Risk Rate')
plt.title('Anxiety Risk Rates among Residents in China')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Anxiety Risk Rates among Residents in China_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
# Display the plot
plt.show()
```
