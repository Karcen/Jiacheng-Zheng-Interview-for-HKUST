```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Given data
years = [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
quantity = [1.00, 13.00, 35.00, 37.00, 25.00, 19.00, 13.00, 8.00, 21.00, 9.00]
amount = [0.48, 4.28, 10.66, 6.76, 5.86, 3.39, 3.19, 5.58, 13.08, 3.87]

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the line chart on the primary y-axis (ax1)
line = ax1.plot(years, quantity, marker='o', label='Quantity', color='blue')
ax1.set_xlabel('Year')
ax1.set_ylabel('Quantity', color='blue')
ax1.tick_params('y', colors='blue')
ax1.legend(loc='upper left')

# Add data labels to the line chart with red color
for i, txt in enumerate(quantity):
    ax1.annotate(f'{txt:.2f}', (years[i], quantity[i]), textcoords="offset points", xytext=(0, 5), ha='center', color='red')

# Create a secondary y-axis (ax2) for the bar chart
ax2 = ax1.twinx()
ax2.bar(years, amount, color='green', alpha=0.5, label='Amount')
ax2.set_ylabel('Amount', color='green')
ax2.tick_params('y', colors='green')
ax2.legend(loc='upper right')

# Add data labels to the bar chart
for i, txt in enumerate(amount):
    ax2.annotate(f'{txt:.2f}', (years[i], amount[i]), textcoords="offset points", xytext=(0, 5), ha='center')

# Set the title
plt.title('Year vs Quantity and Amount')

# Show or save the combined chart
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Year and Quantity combined_chart.png', dpi=1200)
plt.show()
```
