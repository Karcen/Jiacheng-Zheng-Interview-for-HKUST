```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Data
years = ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
events_count = [31, 34, 32, 28, 30, 40, 42, 22]

# Plotting
plt.bar(years, events_count)
plt.xlabel('Year')
plt.ylabel('Event Count')
plt.title('Investment Events in China\'s Mental Health Industry (2015-2022)')
plt.legend(['Event Count'])

# Add labels on top of each bar
for i, v in enumerate(events_count):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Investment Events in Chinas Mental Health Industry (2015-2022)_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
