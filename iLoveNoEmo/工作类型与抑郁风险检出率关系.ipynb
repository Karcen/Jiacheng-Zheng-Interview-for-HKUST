```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'

# Data
job_types = ['Company Employee', 'Technical Professional', 'Manager', 'Civil Servant', 'Business Worker', 'Unemployed', 'Other']
detection_rate = [12.80, 7.50, 3.20, 9.50, 7.00, 31.00, 11.30]

# Plotting
plt.barh(job_types, detection_rate)
plt.ylabel('Job Type')
plt.xlabel('Detection Rate (%)')
plt.title('Relationship between Job Type and Depression Detection Rate')
plt.legend(['Detection Rate'])

# Add labels on right side of each bar
for i, v in enumerate(detection_rate):
    plt.text(v, i, str(v)+'%', ha='left', va='center')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Relationship between Job Type and Depression Detection Rate_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
