```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
reasons = ['High cost', 'Long duration', 'Commuting Time', 'Inefficacy', 'Other']
proportions = [52.90, 7.16, 5.03, 21.95, 12.96]

plt.bar(reasons, proportions, color='orange')

# Add labels to the bars
for i in range(len(reasons)):
    plt.text(reasons[i], proportions[i], str(proportions[i]) + "%", ha='center', va='bottom')

plt.xlabel('Reasons')
plt.ylabel('Proportion (%)')
plt.title('Reasons Why Patients Do Not Consider Psychological Therapy')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Reasons Why Patients Do Not Consider Psychological Therapy_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
