```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Garamond'
# Data
job_status = ['Unemployed', 'Have a job but not working', 'Less working hours than usual', 'More working hours than usual', 'Similar working hours to usual']
detection_rate = [29.80, 27.10, 12.90, 12.80, 6.50]

# Plotting
plt.barh(job_status, detection_rate)
plt.ylabel('Job Status')
plt.xlabel('Detection Rate (%)')
plt.title('Relationship between Job Status and Depression Detection Rate among Chinese Residents')
plt.legend(['Detection Rate'])

# Add labels on right side of each bar
for i, v in enumerate(detection_rate):
    plt.text(v, i, str(v)+'%', ha='left', va='center')
plt.savefig(r'C:\Users\Jiacheng Zheng\Desktop\Re-draw\Relationship between Job Status and Depression Detection Rate among Chinese Residents_en.png', dpi=1200, bbox_inches = 'tight')
plt.show()
```
